```python
import warnings
warnings.filterwarnings('ignore')

# importing modules and packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
covid_df = pd.read_csv('Covid_Data.csv')
covid_df.head()
```

```python
covid_df.dtypes[covid_df.dtypes == "object"].index.tolist()
```

```python
covid_df.shape
```

```python
covid_df = covid_df.drop(covid_df[covid_df["PREGNANT"] > 2].index)
covid_df = covid_df.drop(covid_df[covid_df["PNEUMONIA"] > 2].index)
```

```python
covid_df = covid_df.drop(['MEDICAL_UNIT', "INTUBED","ICU","USMER",'PATIENT_TYPE', 'CLASIFFICATION_FINAL', 'OTHER_DISEASE'], axis = 1)
covid_df.head()
```

```python
covid_df = covid_df.drop(covid_df[covid_df["DATE_DIED"] == '9999-99-99'].index)
```

```python
covid_df
```

```python
covid_df.dtypes
```

```python
Covid_Disease_df = pd.DataFrame(covid_df, columns =['PNEUMONIA','PREGNANT','DIABETES','COPD','ASTHMA','INMSUPR',
                            'HIPERTENSION','CARDIOVASCULAR','OBESITY','RENAL_CHRONIC',
                            'TOBACCO'])
Covid_Disease_df
```

```python
Pneumonia = np.sum(covid_df['PNEUMONIA'] == 1)
print("The number of people who died from Covid AND had pneumonia: ", Pneumonia)
```

```python
Pregnant = np.sum(covid_df['PREGNANT'] == 1)
print("The number of people who died from Covid AND was pregnant: ", Pregnant)
```

```python
Diabetes = np.sum(covid_df['DIABETES'] == 1)
print("The number of people who died from Covid AND had diabetes: ", Diabetes)
```

```python
COPD = np.sum(covid_df['COPD'] == 1)
print("The number of people who died from Covid AND had COPD or Chronic obstructive pulmonary disease: ", COPD)
```

```python
Asthma = np.sum(covid_df['ASTHMA'] == 1)
print("The number of people who died from Covid AND had asthma: ", Asthma)
```

```python
Immunocompromised = np.sum(covid_df['INMSUPR'] == 1)
print("The number of people who died from Covid AND was immunocompromised: ", Immunocompromised)
```
